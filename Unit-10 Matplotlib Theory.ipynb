```python
import matplotlib.pyplot as plt
import numpy as np
# shift+tab on plot

xpoints=np.array([1,8])
ypoints=np.array([3,10])
# plt.plot(xpoints,ypoints,'o') # 'o' gives 2 pts and its known as marker
plt.plot(xpoints,ypoints,marker='H',ms=20) # narker:'o','*','x','X','+','d','D','H'
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
# plt.plot(xpoints,ypoints,'o') # 'o' gives 2 pts and its known as marker
# plt.plot(xpoints,ypoints,marker='o',ms=7,linestyle="--",color="r")
# plt.plot(xpoints,ypoints,'o:r')
plt.plot(xpoints,ypoints,marker='o',ls="dotted",ms='12',mec='r',mfc='y',c='r',lw='4') 
# if x/y-points not given then by-default it will take x-axis as 0,1,2,3
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y1=np.array([3,8,1,10])
y2=np.array([6,2,7,11])
plt.plot(y1,'*:',c='g',ms='25',mec="r",mfc="g")
plt.plot(y2,'d--',c='g',ms='28',mec='b',mfc='m')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y)
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
plt.title("Sport Watch Data",fontdict=font1)
plt.xlabel("Average Pulse",fontdict=font2,loc='right')
plt.ylabel("Calories Burn",fontdict=font2,loc='top')
# plt.grid()
plt.grid(axis='x')
plt.show()
```

```python
# SubPlot

import matplotlib.pyplot as plt
import numpy as np

x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,1,1)
plt.plot(x,y)

x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,1,2)
plt.plot(x,y)

plt.show()

```
